```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D
```

```python
log = pd.read_csv(r'./logs/log.csv')
log.sample(5)
```

```python
fig, ax = plt.subplots()
line1, = ax.plot(log['Epoch'],log['Accuracy'], 'g', label='Accuracy')
line2, = ax.plot(log['Epoch'], log['Loss'], 'r', label='Loss')
ax.legend(handler_map={line1: HandlerLine2D(numpoints=4)})
```
